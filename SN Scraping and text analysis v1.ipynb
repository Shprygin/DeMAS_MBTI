```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer() #по умолччанию русский, для англа в скобочках надо поставить lang='uk'

#теперь мы получили анализатор. Далее нам нужно применить его к каждому слову текста
```

```python
#для примера будем анализировать подпись к одному из моих постов. Я пишу стихи, и вот один из них
text="""
Был высажен мальчиком северным плющ,
Но пора этот куст мне немного подстричь. 
Мой дом стал жертвой токсичных тех гущ,
Секаторам старым я кину свой клич.

Сначала отрежу я добрые ночи,
Затем обрублю плюща плечи, что трогал,
Сколю я финально далёкие очи,
Что плющ обещал, но они лишь из стёкол.

Плющ станет сим новым питомцем в саду.
Где вишни и пихты обрезаны кротко,
Без памяти рыбки играют в пруду.
Крабы, стесняясь, жмутся за лодкой,
У порòся корзинка там яблок в меду.

Обрезанный плющ, наречен мной «безвольным».
Яркий мой сад, всё для жизни бездольной:
Нет места там жару, парам алкогольным.
Там рай мой холодный, лишь мне подконтрольный.

Я обрезал твой плющ. Не трогай! Сломаешь!
Да, теперь это труп, но красивый и мой!
Друг, а гвоздик мне охапку подаришь?
Обрезки живых придают мне покой."""
# note: чтобы str переменная состояла из нескольких строк нужно по три двойных кавычки
```

```python
#теперь надо токенизировать этот текст
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk
nltk.download('punkt')
```

```python
tokenized_text = word_tokenize(text, language="russian") 
tokenized_text
#получили список отдельных слов
```

```python
#проверим, как работает анализаторё
morph.parse(tokenized_text[0])
```

```python
#пример, как мы выводим таг первого слова в списке, где записано много информации
morphalyzed = morph.parse(tokenized_text[0])[0]
print(morphalyzed.tag)
#а теперь выведем только часть речи
print(morphalyzed.tag.POS)
```

```python
#для моей задачи мне нужно считать наречия (adverb), глаголы (verb), существительные (noun), местоимение (pronoun)
#и предлоги (preposition)
from collections import Counter
wordlist=[]
for word in tokenized_text:
    wordlist.append(morph.parse(word)[0].tag.POS)
words_checked=[]
for word in wordlist:
    if word == 'NOUN' or word == 'VERB' or word == 'ADVB' or word == 'NPRO':
        words_checked.append(word)
counts = Counter(words_checked)
print(counts)
# Counter({'apple': 3, 'egg': 2, 'banana': 1})
```
